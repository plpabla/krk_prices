# Import model

```python
import xgboost as xgb
from xgboost import XGBRegressor
import pickle
filename = "../out/krakow_model.pkl"

xgb_reg: XGBRegressor = pickle.load(open(filename, 'rb'))
xgb.plot_importance(xgb_reg, importance_type='weight', max_num_features=10)
```

```python
feature_names = xgb_reg.get_booster().feature_names
print("Feature names:")
for name in feature_names:
    print(name)
```

```python
# extract categorical features
with open("../out/category_mappings_krakow.pkl", "rb") as f:
    category_mappings = pickle.load(f)
    
for key in category_mappings.keys():
    print(f"Key: {key}")
    for i in (category_mappings[key]):
        print(f"  - {category_mappings[key][i]}")
    print("----------------")
```
