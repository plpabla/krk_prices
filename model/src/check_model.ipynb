# Import model

```python
import xgboost as xgb
from xgboost import XGBRegressor
import pickle
filename = "../out/xgboost_model.pkl"

xgb_reg: XGBRegressor = pickle.load(open(filename, 'rb'))
xgb.plot_importance(xgb_reg, importance_type='weight', max_num_features=10)
```

```python
feature_names = xgb_reg.get_booster().feature_names
print("Feature names:")
for name in feature_names:
    print(name)
```

```python
# extract district names from features (starting with location_district)
def get_districts(features: list[str]) -> list[str]:
    return [ feature.split("_")[-1] for feature in features if feature.startswith("location_district") ]

districts = get_districts(feature_names)
print("\nDistricts:")
for district in districts:
    print(district)
```
