# Add luxury level

```python
import pandas as pd

filename = 'otodom.csv'
photos_dir = 'otodom_photos'

df = pd.read_csv(filename)
df.head()
```

```python
import aiofiles.os
import os

async def process_row(df, index):
    row = df.iloc[index]
    slug, state, market = row[['slug', 'state', 'market']].values
    
    if state == 'to_completion' and market == 'primary':
        print(f"Skipping {slug} as it is primary and to_completion")
        df.at[index, 'luxury_level'] = -1
        return
    
    photo_path = os.path.join(photos_dir, slug)
    
    files = await aiofiles.os.listdir(photo_path)
    if not files:
        print(f"No files found in {photo_path}")
        return
    print(f"Processing {len(files)} files in {photo_path}")
```

```python
import asyncio
import nest_asyncio

nest_asyncio.apply()  # Allow nested event loops in Jupyter Notebook

async def process_dataframe(df):
    tasks = []
    for index in range(len(df)):
        tasks.append(process_row(df, index))
    await asyncio.gather(*tasks)

# Run the async processing
await process_dataframe(df)
```
